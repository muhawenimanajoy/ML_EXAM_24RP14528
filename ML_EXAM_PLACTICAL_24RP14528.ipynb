```python
import json
import pandas as pd
```

```python
rows = []

with open("data.txt", "r", encoding="utf-8") as f:
    for line in f:
        try:
            tweet = json.loads(line.strip())
            entities = tweet.get("entities", {})
            
            hashtags = [h.get("text", "") for h in entities.get("hashtags", [])]
            urls = [u.get("expanded_url", "") for u in entities.get("urls", []) if u.get("expanded_url")]
            mentions = [m.get("screen_name", "") for m in entities.get("user_mentions", [])]

            rows.append({
                "created_at": tweet.get("created_at", ""),
                "id": tweet.get("id_str", ""),
                "user": tweet.get("user", {}).get("screen_name", ""),
                "lang": tweet.get("lang", ""),
                "text": tweet.get("text", ""),
                "hashtags": ",".join([str(h) for h in hashtags if h]),
                "urls": ",".join([str(u) for u in urls if u]),
                "mentions": ",".join([str(m) for m in mentions if m])
            })
        except json.JSONDecodeError:
            continue

df = pd.DataFrame(rows)

print(f"Saved {len(df)} tweets with entities to tweets_dataset.csv")
```

# For now, let's rename it properly

```python
df.rename(columns={'user': 'username'}, inplace=True)
```

```python
df.head(5)
```

# Replace empty strings with NaN

```python
import numpy as np
df.replace("", np.nan, inplace=True)
```

# Drop duplicates (if any)

```python
df.drop_duplicates(inplace=True)
```

# Clean datetime

```python
df['created_at'] = pd.to_datetime(
    df['created_at'],
    format="%a %b %d %H:%M:%S %z %Y",
    errors="coerce"
)
```

```python
df.head(5)
```

# Split comma-separated hashtags into lists

```python
df['hashtags'] = df['hashtags'].dropna().apply(lambda x: [h.strip() for h in str(x).split(",") if h.strip() != ""])
```

# Split mentions

```python
df['mentions'] = df['mentions'].dropna().apply(lambda x: [m.strip() for m in str(x).split(",") if m.strip() != ""])
```

# Split urls

```python
df['urls'] = df['urls'].dropna().apply(lambda x: [u.strip() for u in str(x).split(",") if u.strip() != ""])
```

# Fill missing structured columns

```python
df['hashtags'] = df['hashtags'].apply(lambda x: x if isinstance(x, list) else [])
df['mentions'] = df['mentions'].apply(lambda x: x if isinstance(x, list) else [])
df['urls'] = df['urls'].apply(lambda x: x if isinstance(x, list) else [])
```

# Derived features from tweet timestamp

```python
df['tweet_year']   = df['created_at'].dt.year
df['tweet_month']  = df['created_at'].dt.month
df['tweet_day']    = df['created_at'].dt.day
df['tweet_hour']   = df['created_at'].dt.hour
df['tweet_weekday']= df['created_at'].dt.weekday  # 0=Mon, 6=Sun
df['is_weekend']   = df['tweet_weekday'].isin([5,6]).astype(int)
```

# Account age (in days) at time of tweet

```python
if 'user_created_at' in df.columns:
    df['account_age_days'] = (df['created_at'] - df['user_created_at']).dt.days
```

```python
import re
```

```python
df['clean_text'] = df['text'].str.lower()
```

# Remove URLs, mentions, hashtags, special chars

```python
df['clean_text'] = df['clean_text'].apply(
    lambda x: re.sub(r"http\S+|www\S+|https\S+", "", x)
)
df['clean_text'] = df['clean_text'].apply(
    lambda x: re.sub(r"@\w+|#\w+", "", x)
)
df['clean_text'] = df['clean_text'].apply(
    lambda x: re.sub(r"[^a-zA-Z0-9\s]", "", x)
)
```

# Text-based features

```python
df['char_count']  = df['text'].apply(len)
df['word_count']  = df['text'].apply(lambda x: len(x.split()))
df['avg_word_len']= df['text'].apply(lambda x: sum(len(w) for w in x.split())/len(x.split()) if len(x.split())>0 else 0)
```

# Assumptions & Choices

### - Dropped duplicates this will prevents model bias.  
### - Dropped missing text/timestamps ensurure that text and time are essential.  
### - Account age & tweet frequency → help distinguish bots from genuine users.  
### - Text normalization (lowercase, remove URLs/mentions/hashtags) → avoids noise in NLP models.  
### - Derived flags (has_url, has_mention, has_hashtag) → capture spam-like behavior.

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
```

df.info()

```python
X = df["clean_text"]
y = df["label"]
```

```python
vectorizer = TfidfVectorizer(max_features=5000, stop_words="english")
X_tfidf = vectorizer.fit_transform(X)
```

```python
# Train Logistic Regression model
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.2, random_state=42)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```
